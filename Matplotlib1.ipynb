```python
x = np.linspace(0,10,100)
fig = plt.figure() # define the properties of figure
plt.plot(x , np.sin(x) , '-')
plt.plot(x , np.cos(x) , '--')
```

```python
# Two Interfaces for the Price of One
# MATLAB-STYLE INTERFACE
plt.figure()  # create a plot figure
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
# plt.gcf() get current figure
# plt.gca() get current axes 
```

```python
# OBJECT-ORIENTED INTERFACE for more complicated situations
# First create a grid of plots
fig,ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# Adjusting the Plot: Line Colors and Styles
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF) 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) #  RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse') # # all HTML color names supported
```

```python
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':') # dotted 
# we can also use 'solid' directly 
```

```python
# linestyle and color codes can be combined
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r') # dot red
```

```python
# Adjusting the Plot: Axes Limits plt.xlim() plt.ylim()
plt.plot(x , np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
# reverse the display of axis
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
# The plt.axis() method allows you to set the x and y limits with a single call
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
# plt.axis() can do more than this
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# Labeling Plots
plt.plot(x, np.sin(x), '--g')
plt.axis([0,10,-1.5,1.5])
plt.title('My curve')
plt.xlabel('X')
plt.ylabel('Sin(X)') 
# u can adjust the position, size and style of these labels using optional arguments
```

```python
# when multiple lines are created in a single axes, use plt.legend() to create plot legend 
plt.plot(x , np.sin(x), '-g' , label= 'sin(X)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

```python
# in object-oriented interface to plotting, use ax.Set() to set all these properties at once
ax = plt.axes()
ax.plot(x , np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),xlabel='x', ylabel='sin(x)',title='A Simple Plot')
```

```python
# Simple Scatter Plots with plt.plot or ax.plot 
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=f"marker = {marker}")
plt.legend(numpoints = 1)
plt.xlim(0, 1.8)
```

```python
plt.plot(x , y , '-ok') # # line (-), circle marker (o), black (k)
```

```python
plt.plot(x, y, '-p', color='gray',markersize=15, linewidth=4,
         markerfacecolor='red', markeredgecolor='green', markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
# plt.scatter more powerful to create scatter , this function is better to control the properties of individual points
plt.scatter(x , y , marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x , y , c=colors , s=sizes , alpha=0.3, cmap='viridis') # color maps 
plt.colorbar() # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
# visualizing errors
# basic errorbars
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0)
```
