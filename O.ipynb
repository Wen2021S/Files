```python
# KNN classifier 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
np.unique(y)
```

```python
def mytraintestsplit(x,y,test_size):
    idx = np.random.permutation(np.arange(x.shape[0]))
    ts = int(np.floor(test_size*x.shape[0]))
    testidx=idx[:ts]
    trainidx=idx[ts:]
    trainx=x[trainidx,:]
    trainy=y[trainidx]
    testx =x[testidx,:]
    testy=y[testidx]
    return trainx,trainy,testx,testy
```

```python
trainx,trainy,testx,testy = mytraintestsplit(x,y,0.2)
```

```python
D = np.sum((trainx[:,np.newaxis,:]-testx[np.newaxis,:,:])**2,axis=2)
```

```python
nidx=np.argsort(D,axis=1)
```

```python
nidx[0,:5]
```

```python
idx =nidx[:5,0]
```
